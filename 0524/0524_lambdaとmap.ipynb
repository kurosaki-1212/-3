# **lambdaの例**

```python
# 引数のnを3倍にする関数
def triple(n):
    return n * 3

data = 10

# 関数を使う例
result1 = triple(data)

# 引数のnを3倍にするlambda
result2 = lambda n : n * 3
print(result1)

# lambdaを使う例
print(result2(data))
```

# **mapの例**

```python
a = '11,12,13'
# ①a.split(',')→aの内容を , で区切ってそれぞれに分割する
# ②mapで int型に変換し、x,y,zにそれぞれ代入する 
x,y,z =  map(int,a.split(','))

sum = x + y + z
sum
```

# **mapとlambdaを組み合わせて利用する例**

```python
# ①map も lambdaも使わない例

# 元データ
test = [1, 2, 3]
result3 = [] # 足し算の結果を入れるリスト

for i in test:
    result3.append(i + 5)  # 5 を足した結果を追加

print(result3)  # [6, 7, 8]
```

```python
# ②mapを使う例

def add_five(m):
    return m + 5

# 元データ
test2 = [1, 2, 3]

result4 = map(add_five, test2)

print(list(result4))  # [6, 7, 8]
```

```python
# ③map も lambdaも使う例
test3 = [1, 2, 3]
result5 = map(lambda i: i + 5, test3)
print(list(result5))  # [6, 7, 8]
```

# **練習①：mapとlambdaを使って、リストの中身を10倍にしリスト型にしてみよう**

```python
test4 = [1, 2, 3, 4, 5]
result6 = map(lambda i: i * 10, test4)
print(list(result6))  # [10, 20, 30, 40, 50]
```

# **練習②：mapとlambdaを使って、リストの中身をインチに変換しリスト型にしてみよう**

```python
# ヒント：センチデータからインチデータに変換する場合は、2.54を掛ける

test5 = [1, 2, 3, 4, 5]
result7 = map(lambda i: i * 2.54, test5)
print(list(result7))
```

# **練習③：mapを使って、文字列の中身を「,」で分割し、int型に変換しリスト型にしてみよう**

```python
test6 = '1, 2, 3, 4, 5'
result8 = map(int, test6.split(','))
print(list(result8))
```
