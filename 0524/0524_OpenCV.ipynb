```python
# ①画像のダウンロード
import urllib.request as req
url = "https://uta.pw/shodou/img/28/214.png"
req.urlretrieve(url, "test.png")

# OpenCVで読み込む
import cv2
img = cv2.imread("test.png")
print(img)
```

```python
# ダウンロードした画像を画面に表示する
import matplotlib.pyplot as plt
import cv2
img_yakitori = cv2.imread("test.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 画像のサイズを表示
img.shape

# 縦サイズ, 横サイズ, チャネルの順で表示されている
```

```python
# 自分の好きな画像データをAWFW2フォルダに格納し、表示してみよう
import matplotlib.pyplot as plt
import cv2
img_kinniku = cv2.imread("kinniku.jpg")
plt.imshow(cv2.cvtColor(img_kinniku, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#  axisの表示をオフにして表示してみましょう
plt.axis('off')
plt.imshow(cv2.cvtColor(img_kinniku, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 読み込んだ画像を「out.jpg」として保存しましょう
cv2.imwrite('out.jpg', img)
```

```python
# 読み込んだ画像のサイズを表示しましょう
img_kinniku.shape
```

```python
# 読み込んだ画像を縦に2倍引き延ばして表示しましょう
im2_kinniku = cv2.resize(img_kinniku, (225, 550))
plt.imshow(cv2.cvtColor(im2_kinniku, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 画像の最も重要な部分と思われる箇所を切り出して表示しましょう
im3_kinniku = img_kinniku[0:100, 80:150]
im3_kinniku = cv2.resize(im3_kinniku, (225, 225))
plt.imshow(cv2.cvtColor(im3_kinniku, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 画像をグレースケールに変換しましょう
# ヒント：P135 
plt.imshow(cv2.cvtColor(im3_kinniku, cv2.COLOR_BGR2GRAY))
plt.show()
```

```python
# 復習１
# ①で読み込んだ「やき鳥」の画像の「鳥」の文字に赤い四角形を描画し、表示しましょう。
# ヒント：rectangle
img_yakitori = cv2.rectangle(img_yakitori,
               pt1=(390, 510),
               pt2=(200, 290),
               color=(0, 0, 255),
               thickness=3,
               lineType=cv2.LINE_4,
               shift=0)
plt.imshow(cv2.cvtColor(img_yakitori, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 復習２
# 復習１で赤い四角形を描画した描画したデータにさらに「Delicious!」という文字を重ねましょう。
# 文字を重ねる位置はどこでもOKです。文字色も自由に設定してください。
# ヒント：putText

im2_yakitori = cv2.putText(img_yakitori,
               text='Delicious!',
               org=(100, 300),
               fontFace=cv2.FONT_HERSHEY_SIMPLEX,
               fontScale=2.0,
               color=(0, 0, 255),
               thickness=4,
               lineType=cv2.LINE_4)

plt.imshow(cv2.cvtColor(im2_yakitori, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# 復習２で作成した画像を「test2.png」という名前で保存しましょう
# 画像を保存する
cv2.imwrite('test2.png', img)
```
